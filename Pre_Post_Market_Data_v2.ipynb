```python
# Convert to correct Data Types

# Convert Date to datetime
df_input["DATE"] = pd.to_datetime(df_input["DATE"])


# Convert Market Session column to string
df_input["Market_Session"] = df_input["Market_Session"].astype(str)

```

```python
# --- More Efficient Aggregation ---
df_filtered = df_input[df_input["Market_Session"].isin(["Pre-Market", "Post-Market"])]

daily_df = df_filtered.pivot_table(index=["SYM_ROOT", "DATE"],
                                   columns="Market_Session",
                                   values="SIZE",
                                   aggfunc="sum").reset_index()

# Rename columns for clarity
daily_df.columns = ["SYM_ROOT", "DATE", "pre_market_volume", "post_market_volume"]
```

```python
# Pre-filter for Pre-Market and Post-Market trades
pre_market_df = df_input[df_input["Market_Session"] == "Pre-Market"]
post_market_df = df_input[df_input["Market_Session"] == "Post-Market"]

# Aggregate separately
pre_market_agg = pre_market_df.groupby(["SYM_ROOT", "DATE"])["SIZE"].sum().rename("pre_market_volume")
post_market_agg = post_market_df.groupby(["SYM_ROOT", "DATE"])["SIZE"].sum().rename("post_market_volume")

# Merge the two dataframes
daily_df = pd.merge(pre_market_agg, post_market_agg, on=["SYM_ROOT", "DATE"], how="outer").reset_index()
```

```python
# --- Aggregate Data at the Daily Level ---
daily_df = df_input.groupby(["SYM_ROOT", "DATE"]).agg(
    pre_market_volume=("SIZE", lambda x: x[df_input["Market_Session"] == "Pre-Market"].sum()),
    post_market_volume=("SIZE", lambda x: x[df_input["Market_Session"] == "Post-Market"].sum()),
).reset_index()
```

```python
# Price features: mean, std, min, max, overnight squared return
price_features = df_input.groupby(["SYM_ROOT", "DATE"]).agg(
    PRICE_MEAN=("PRICE", "mean"),
    PRICE_STD=("PRICE", "std"),
    PRICE_MIN=("PRICE", "min"),
    PRICE_MAX=("PRICE", "max"),
    PRICE_MEDIAN=("PRICE", "median")
).reset_index()
```

```python
# Convert 'TIME_M' to datetime to calculate the difference
df_input["TIME_M"] = pd.to_datetime(df_input["TIME_M"])
```

```python
# Filter for Pre-market (preopen) and Post-market (postclose) trades
pre_market_df = df_input[df_input["Market_Session"] == "Pre-Market"]
post_market_df = df_input[df_input["Market_Session"] == "Post-Market"]

# Merge the Pre-market and Post-market for each symbol and date
pre_post_df = pd.merge(pre_market_df, post_market_df, on=["SYM_ROOT", "DATE"], how="outer", suffixes=("_pre", "_post"))

# Calculate overnight return as the price difference between the first pre-open and last post-close
pre_post_df["overnight_return"] = pre_post_df["PRICE_post"] - pre_post_df["PRICE_pre"]

# Calculate squared overnight return
pre_post_df["overnight_squared_return"] = pre_post_df["overnight_return"] ** 2
```

```python
# Aggregate overnight squared return by symbol and date
overnight_return_agg = pre_post_df.groupby(["SYM_ROOT", "DATE"])["overnight_squared_return"].sum().rename("overnight_squared_return").reset_index()

# Merge price features with pre-market and post-market volume
daily_df = pd.merge(daily_df, price_features, on=["SYM_ROOT", "DATE"], how="outer")
daily_df = pd.merge(daily_df, overnight_return_agg, on=["SYM_ROOT", "DATE"], how="outer")

# Display the final dataframe with added price features
import ace_tools as tools
tools.display_dataframe_to_user(name="Daily Aggregated Price and Volume Features", dataframe=daily_df)
```

```python
# Add 'DAY_OF_WEEK' column
daily_df['DAY_OF_WEEK'] = daily_df['DATE'].dt.day_name()
```

# Extracting the Target Variables from the OHVLC Data.

```python
# Initialize an empty DataFrame for merging
final_dataset = daily_df.copy()

# Filter for dates up to '2019-04-30'
end_date = '2020-02-29'

# Initialize an empty DataFrame to consolidate all target variable data
target_data_combined = pd.DataFrame()

# Loop through OHVLC files and extract relevant labels
for file in output_files:
    print(f"Processing file: {file}")

    # Load OHVLC file
    try:
        ohlcv_data = pd.read_csv(file, skiprows=3, names=['DATE', 'CLOSE', 'HIGH', 'LOW', 'OPEN', 'VOLUME', 'ADJ CLOSE'])
    except Exception as e:
        print(f"Error loading file {file}: {e}")
        continue

    if ohlcv_data.empty:
        print(f"OHVLC data is empty for {file}")
        continue

    # Ensure DATE is in datetime format
    ohlcv_data['DATE'] = pd.to_datetime(ohlcv_data['DATE'])

    # Filter rows up to the end date
    ohlcv_data = ohlcv_data[ohlcv_data['DATE'] <= end_date]

    # Extract ticker from the filename
    ticker = os.path.basename(file).split("_")[0]
    ohlcv_data['SYM_ROOT'] = ticker

    # Extract necessary columns
    ohlcv_data = ohlcv_data[['DATE', 'SYM_ROOT', 'OPEN', 'HIGH', 'LOW', 'CLOSE', 'VOLUME']]

    # Append to consolidated DataFrame
    target_data_combined = pd.concat([target_data_combined, ohlcv_data], ignore_index=True)


# Merge labels into your dataset
final_dataset = pd.merge(final_dataset, target_data_combined, on=['DATE', 'SYM_ROOT'], how='left')

# Save the final dataset
output_with_labels_path = r"/content/drive/MyDrive/Dataset/Final_Dataset/final_dataset_with_labels_v4.csv"
final_dataset.to_csv(output_with_labels_path, index=False)
print(f"Dataset with labels saved to: {output_with_labels_path}")
```
